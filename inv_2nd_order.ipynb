```python
import pandas as pd

df = pd.read_csv('data/jj.csv')

print(len(df))
df.head()
```

```python
import numpy as np

eps_diff = np.diff(df['data'], n=2)
```

```python
N_POINTS = len(df)

y_0, y_1 = df['data'].iloc[:2]

eps_restored = np.empty(N_POINTS)
eps_restored[:2] = [y_0, y_1]

for n in range(2, N_POINTS):
    val = n * y_1 - (n - 1) * y_0
    for i in range(2, n+1):
        val += (n - i + 1) * eps_diff[i-2]
    eps_restored[n] = val
    
df['data_restored'] = eps_restored
df
```
